```python
# Define colors for each class (BGR format for OpenCV)
CLASS_COLORS = {
    0: (255, 0, 0),     # Class 0 - Blue
    1: (0, 255, 0),     # Class 1 - Green
    2: (0, 0, 255),     # Class 2 - Red
}
```

```python
import os
import cv2
from ultralytics import YOLO

# -----------------------------
# Paths
# -----------------------------
MODEL_PATH = "/kaggle/input/enhanced-yolov8-l-on-seadronessee/pytorch/v1/6/last.pt"
IMAGE_DIR = "/kaggle/input/inference-sample/Images"
OUTPUT_DIR = "/kaggle/working/inferenced_images"

os.makedirs(OUTPUT_DIR, exist_ok=True)

# -----------------------------
# Load model
# -----------------------------
model = YOLO(MODEL_PATH)

# Class names from model
CLASS_NAMES = model.names

# -----------------------------
# Legend drawing function
# -----------------------------
def draw_legend(img, class_names, class_colors):
    x_start, y_start = img.shape[1] - 200, img.shape[0] - (30 * len(class_names)) - 10
    box_size = 15

    for i, (cls_id, cls_name) in enumerate(class_names.items()):
        color = class_colors.get(cls_id, (255, 255, 255))
        y = y_start + i * 30

        # Color box
        cv2.rectangle(
            img,
            (x_start, y),
            (x_start + box_size, y + box_size),
            color,
            -1
        )

        # Class text
        cv2.putText(
            img,
            cls_name,
            (x_start + 25, y + 13),
            cv2.FONT_HERSHEY_SIMPLEX,
            0.5,
            (255, 255, 255),
            1,
            cv2.LINE_AA
        )

# -----------------------------
# Run inference on folder
# -----------------------------
results = model.predict(
    source=IMAGE_DIR,
    conf=0.25,
    iou=0.45,
    imgsz=640,
    stream=True   # memory efficient
)

for r in results:
    img = r.orig_img.copy()

    if r.boxes is not None:
        for box, cls_id in zip(r.boxes.xyxy, r.boxes.cls):
            x1, y1, x2, y2 = map(int, box)
            cls_id = int(cls_id)

            color = CLASS_COLORS.get(cls_id, (255, 255, 255))

            # Draw bounding box ONLY (no text)
            cv2.rectangle(img, (x1, y1), (x2, y2), color, 2)

    # Draw legend
    draw_legend(img, CLASS_NAMES, CLASS_COLORS)

    # Save image
    output_path = os.path.join(OUTPUT_DIR, os.path.basename(r.path))
    cv2.imwrite(output_path, img)

print("✅ Inference completed with color-coded boxes and legend")
```
